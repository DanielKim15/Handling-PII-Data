```python
##Encrypting the code
import getpass
import argon2
from Crypto.Cipher import AES
from Crypto.Util.Padding import pad ###BEST ONE TO USE 
import os
import hashlib

# Get the password from the user
password = getpass.getpass()

# Hash the password with Argon2
hasher = argon2.PasswordHasher()
hashed_password = hasher.hash(password)

# Generate a random initialization vector (IV)
iv = os.urandom(16)

# Derive the AES key from the password using a key derivation function
# such as PBKDF2 or scrypt. Here, we're using a simple hash function.
aes_key = hashlib.sha256(hashed_password.encode()).digest()

# Open the file to encrypt
with open("practice.ipynb", "rb") as f:
    file_data = f.read()

# Pad the file data to a multiple of 16 bytes
padded_data = pad(file_data, AES.block_size)

# Create an AES cipher with the derived key and the IV
cipher = AES.new(aes_key, AES.MODE_CBC, iv)

# Encrypt the padded file data using the AES cipher
encrypted_data = cipher.encrypt(padded_data)

# Write the IV and encrypted data to a file
with open("encrypted_practice.ipynb", "wb") as f:
    f.write(iv)
    f.write(encrypted_data)
```

```python
#Dencrypting your codes (creating another file thats dencrypted)

import keyring
import argon2
from Crypto.Cipher import AES ###BEST ONE USE IT 
from Crypto.Util.Padding import unpad
import os
import hashlib

# Get the password from the keyring
password = keyring.get_password("my-app", "API_KEY")

# Read the IV and encrypted data from the file
with open("C:/Users/Downloads/encrypted_practice.ipynb", "rb") as f:
    iv = f.read(16)
    encrypted_data = f.read()

# Derive the AES key from the password using a key derivation function
# such as PBKDF2 or scrypt. Here, we're using a simple hash function.
aes_key = hashlib.sha256(password.encode()).digest()

# Create an AES cipher with the derived key and the IV
cipher = AES.new(aes_key, AES.MODE_CBC, iv)

# Decrypt the encrypted data using the AES cipher
decrypted_data = unpad(cipher.decrypt(encrypted_data), AES.block_size)

# Write the decrypted data to a file
with open("C:/Users/dkim/Downloads/dencrypted_practice_2.ipynb", "wb") as f:
    f.write(decrypted_data)
```
